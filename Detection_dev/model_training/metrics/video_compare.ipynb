```python
from ultralytics import YOLO

# 1. Define paths to the models you want to compare
base_model_path = "yolov8n.pt"  # Base, pre-trained YOLO model
v3_model_path = r"runs\detect\Modele_Carla\osobowki_model_v3\weights\best.pt" # Your custom V3 model

# Path to the input video file
video_source = r"../yolo8_v0/data/carla_traffic.mp4"

def process_video(model_path, source_video):
    """
    Loads a YOLO model and processes an input video, saving the annotated result.
    """
    print(f"\nLoading model: {model_path}...")
    model = YOLO(model_path)
    
    print("Starting video processing...")
    # The predict function automatically handles video streams frame by frame
    model.predict(
        source=source_video,
        save=True,       # CRITICAL: Saves the resulting annotated video to disk
        conf=0.5,        # Confidence threshold. Shows only predictions with >= 50% confidence
        show=True,       # Set to True to see a live preview during rendering
    )
    print("Processing complete! Check the 'runs/detect/predict' folder.")

if __name__ == "__main__":
    # Run the video test for the base YOLOv8 model
    print("--- BASE MODEL TEST ---")
    process_video(base_model_path, video_source)
    
    # Run the video test for your custom V3 model
    print("\n--- CUSTOM V3 MODEL TEST ---")
    process_video(v3_model_path, video_source)
```
