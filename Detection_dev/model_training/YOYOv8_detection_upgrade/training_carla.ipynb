```python
from ultralytics import YOLO

# upload lightweight model
model = YOLO('yolov8n.pt')

print("Training will start now. This may take a while...")

# begin training
results = model.train(
    data='carla-osobowki-przejscie-1/data.yaml', # Podmień na rzeczywistą nazwę folderu z drzewa plików!
    epochs=150,       # 150 epochs to allow for better learning, adjust as needed
    patience=30,      # Early stopping patience to prevent overfitting
    imgsz=416,        # size of input images, 416 is a common choice for YOLO models
    batch=16,         # Value of 16 to fit within 6GB VRAM, adjust if you encounter out-of-memory errors
    device=0,         # Use GPU (0 for the first GPU)
    cache=True,       # Wrzuca dataset do RAM, przyspieszając trening
    project="Modele_Carla",
    name="osobowki_model_v3"
)

print("\nTraining completed! Your weights (best.pt) can be found in the folder Modele_Carla/osobowki_model_v3/weights/")
```
