```python
import os
from dotenv import load_dotenv
from roboflow import Roboflow


load_dotenv()

api_key = os.getenv("ROBOFLOW_API_KEY")

if not api_key:
    raise ValueError("APIKEY not found")

# Inicjalizacja Roboflow
rf = Roboflow(api_key=api_key)

# Pobieranie projektu (używamy v19 i poprawnej nazwy formatu yolo26)
project = rf.workspace("archie-junio-dxv5t").project("car-detection-model-bwjpb")
dataset = project.version(19).download("yolo26")

print(f"Sukces! Dane pobrane do: {dataset.location}")
```

```python
from ultralytics import YOLO


model = YOLO("yolo26n.pt")

model.train(
    data=f"{dataset.location}/data.yaml", 
    epochs=50, 
    imgsz=640,
    name="yolo_car",  
    save=True,             
    save_period=10          
)
```
